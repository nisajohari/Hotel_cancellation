# Hotel booking data prediction and cancellation

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
raw_csv_data = pd.read_csv("hotel_bookings.csv")
df = raw_csv_data.copy()
pd.set_option('display.max_rows',2000,'display.max_columns',40)
df
```

```python
df.describe()
```

Dataset: contain 119390 observation with 32 variables such as status reservatio, lead time, arrival date, stays in week nights, adults, children, babies, meal, country and etc.
Max number of lead time is 737 days with mean 104 days.


## Data preparation

```python
(df.isnull().sum()/len(df))*100
```

Percentage of null value on column company is 94.31%. So we going to drop that company column from a DataFrame

```python
df = df.drop(['company'], axis = 1)
df
```

```python
for col in df.columns:
    if df[col].dtype == 'object' and col != 'country':
        df.loc[df[col].str.contains('NULL'), col] = np.nan
        df.loc[df[col].str.contains('Undefined', na=False), col] = np.nan
null_series = df.isnull().sum()
print(null_series[null_series > 0])
```

```python
subset = ['country','children','market_segment','distribution_channel'] 
df = df.dropna(subset=subset)
```

ADR refer to average daily rate, based on describe() above, the minimum value of ADR is -6.38 which is wrong value. We delete values that are smaller than 0

```python
df = df[df.adr > 0]
df.describe()
```

Fill in the empty value with NIL for country, 0.0 for agent and 0.0 for children. Consider meal value of 'undefined' as SC(self catering).

```python
df['country'] = df['country'].fillna (value='NIL')
df['agent'] = df['agent'].fillna (value=0.0)
df['children'] = df['children'].fillna (value=0.0)
df.loc[df.meal.isnull(), 'meal'] = 'SC'
```

```python
df['children']= df['children'].astype(int)
df
```

Deleting **zero** total number of adult, child, and babies.

```python
filter = (df.children == 0) & (df.adults == 0) & (df.babies == 0)
df = df[~filter]
df
```

## Descriptive statistics

```python
months = ["January", "February", "March", "April", "May", "June", 
          "July", "August", "September", "October", "November", "December"]
df['arrival_date_month'] = pd.Categorical(df['arrival_date_month'], categories=months, ordered=True)
df = df.sort_values(by="arrival_date_month")
checkin_guests = df.loc[df["is_canceled"] == 0] 
plt.figure(figsize=(15,20))
sns.countplot(x='hotel',hue='arrival_date_month',data=checkin_guests)
plt.title("Check-in guest within Hotel by Months", fontsize=16)
plt.xlabel("Month", fontsize=13)
plt.ylabel("Sum of Guest", fontsize=13)
```

**When is the peak time and low time?** Most crowded time of the year where guest check-in is on July and August in both hotel. Low occupancy rate during November until January due to winter season. Hotel can give more discount during low season to attract guest.

```python
df['nonights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
no_nights_max = df.loc[df["is_canceled"] == 0]
no_nights_max.describe()
```

```python

plt.figure(figsize=(15,20))
sns.countplot(x='nonights',hue='hotel',data=no_nights_max[-(no_nights_max.nonights>20)])
plt.title("Check-in guest's number of nights stay by hotels", fontsize=16)
plt.xlabel("Number of nights", fontsize=13)
plt.ylabel("Sum of bookings", fontsize=13)
```

**How many nights guest stay in each hotel?** For city hotel guest stay for short term which is 1-4 nights, while the resort hotel guest usually stay for 1 night and decreasing until 6 nights, but suprisingly many guest stay for 7 nights. Maybe due to 1 week(7 nights) package or discount that the resort hotel offered.

```python
is_cancel_count = (df.groupby(['is_canceled'])['hotel']
                     .value_counts(normalize=True)
                     .rename('percentage')
                     .mul(100)
                     .reset_index()
                     .sort_values('hotel'))
plt.figure(figsize=(8,6))
sns.barplot(x='hotel',y='percentage',hue='is_canceled',data=is_cancel_count)
plt.title("Percentage of canceled by hotel", fontsize=16)
plt.xlabel("Hotel", fontsize=13)
plt.ylabel("Percentage of reservation", fontsize=13)
```

**What is the percentage of cancellation in the hotels?** the percentage of guest canceled reservation at city hotel is higher than guest who check in. While at resort hotel, guest who check in is higher than guest that canceled reservation 

```python
cancel_corr = df.corr()["is_canceled"]
cancel_corr.abs().sort_values(ascending=False)[1:]
```

## Create features

```python
df_subset = df.copy()
```

```python
#more_canceled if previous_cancellations more than previous_bookings_not_canceled 1 else 0
df_subset['more_canceled'] = 0
df_subset.loc[ df_subset['previous_cancellations'] > df_subset['previous_bookings_not_canceled'] , 'more_canceled'] = 1

#Room_choice if assign room same 1, else 0
df_subset['Room_choice'] = 0
df_subset.loc[ df_subset['reserved_room_type'] == df_subset['assigned_room_type'] , 'Room_choice'] = 1

df_subset['reservation_status_date'] = pd.to_datetime(df_subset['reservation_status_date'])
df_subset['reservation_year'] = df_subset['reservation_status_date'].dt.year
df_subset['reservation_month'] = df_subset['reservation_status_date'].dt.month
df_subset['reservation_day'] = df_subset['reservation_status_date'].dt.day

## remove the leftover columns. 
df_subset = df_subset.drop(['reservation_status'], axis=1)
df_subset = df_subset.drop(['booking_changes','assigned_room_type','reservation_status_date','distribution_channel'],axis=1)
df_subset
```

Transform categorical variables into numbering variable

```python
cat_col = ['hotel', 'arrival_date_month','meal','country','market_segment','reserved_room_type',
               'deposit_type','agent','customer_type']

def transform(dataframe):
    
    from sklearn.preprocessing import LabelEncoder
    
    le = LabelEncoder()
    dataframe[cat_col]=dataframe[cat_col].apply(lambda x: le.fit_transform(x))
    return dataframe

df_subset = transform(df_subset)
df_subset
```

```python
df_subset.var()
```

```python
df_subset['lead_time']=np.log(df_subset['lead_time']+1)
df_subset['agent'] = np.log(df_subset['agent'] + 1)
df_subset['adr'] = np.log(df_subset['adr'] + 1)
df_subset['country'] = np.log(df_subset['country'] + 1)
df_subset['days_in_waiting_list'] = np.log(df_subset['days_in_waiting_list'] + 1)
df_subset['reservation_month'] = np.log(df_subset['reservation_month'] + 1)
df_subset['reservation_day'] = np.log(df_subset['reservation_day'] + 1)
df_subset['arrival_date_week_number'] = np.log(df_subset['arrival_date_week_number'] + 1)
df_subset['arrival_date_day_of_month'] = np.log(df_subset['arrival_date_day_of_month'] + 1)
df_subset['arrival_date_month'] = np.log(df_subset['arrival_date_month'] + 1)
df_subset['nonights'] = np.log(df_subset['nonights'] + 1)

df_subset = df_subset.drop([])
```

```python
df_subset.var()
```

Split the data into test and train data sets -20%, 80% respectively

```python
def data_split(df, label):
    
    from sklearn.model_selection import train_test_split
    
    X = df.drop(label, axis=1)
    Y = df[label]
    x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size = 0.20)    
    return x_train, x_test, y_train, y_test

x_train, x_test, y_train, y_test = data_split(df_subset, 'is_canceled')

from sklearn.impute import SimpleImputer

imputer = SimpleImputer()
feature_names = x_train.columns
x_train = imputer.fit_transform(x_train)
x_test = imputer.transform(x_test)
```

```python
feature_names
```

## Statistics

Model used for training: LogisticRegression, KNeighborsClassifier, XGBClassifier and CatBoostClassifier

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

lr = LogisticRegression(solver='liblinear')
lr.fit(x_train, y_train)

pred_lr = lr.predict(x_test)
clf_report = classification_report(y_test, pred_lr)
print(f"Classification Report : \n{clf_report}")

lr_acc = accuracy_score(pred_lr, y_test)
print('Logistic Regression Accuracy Score: ', lr_acc)
```

```python
lr_conf = confusion_matrix(pred_lr, y_test)
lr_confusion = pd.DataFrame(data=lr_conf, columns=[['Predicted Not Canceled', 'Predicted Canceled']],
                           index = [['Actual Not Canceled', 'Actual Canceled']])
fig, ax = plt.subplots(figsize = (8,8))
sns.heatmap(lr_confusion, annot=True, fmt='g')
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)

pred_knn = knn.predict(x_test)

clf_report_knn = classification_report(y_test, pred_knn)
print(f"Classification Report : \n{clf_report_knn}")
knn.score(x_test,y_test)

knn_acc = accuracy_score(pred_knn, y_test)
print('KNeighborsClassifier Accuracy Score: ', knn_acc)
```

```python
knn_conf = confusion_matrix(pred_knn, y_test)
knn_confusion = pd.DataFrame(data=knn_conf, columns=[['Predicted Not Canceled', 'Predicted Canceled']],
                           index = [['Actual Not Canceled', 'Actual Canceled']])
fig, ax = plt.subplots(figsize = (8,8))
sns.heatmap(knn_confusion, annot=True, fmt='g')
```

```python

xgb = XGBClassifier(booster = 'gbtree', learning_rate = 0.1, max_depth = 5, n_estimators = 500,use_label_encoder=False)
xgb.fit(x_train, y_train)

pred_xgb = xgb.predict(x_test)

clf_report_xgb = classification_report(y_test, pred_xgb, digits=5)
print(f"Classification Report : \n{clf_report_xgb}")

xgb_acc = accuracy_score(pred_xgb, y_test)
print('XGBClassifier Accuracy Score: ', xgb_acc)
```

```python
xgb_conf = confusion_matrix(pred_xgb, y_test)
xgb_confusion = pd.DataFrame(data=xgb_conf, columns=[['Predicted Not Canceled', 'Predicted Canceled']],
                           index = [['Actual Not Canceled', 'Actual Canceled']])
fig, ax = plt.subplots(figsize = (8,8))
sns.heatmap(xgb_confusion, annot=True, fmt='g')
```

**CatBoost** There are 4 observation classified as not canceled when they suppose to be canceled. Meanwhile 10 observation classified as canceled while they actually not canceled.

**Accuracy = 99.94%**

**Precision = 99%**

**Recall= 99%**

```python
clf = DecisionTreeClassifier(random_state=0)
clf.fit(x_train,y_train)
pred_clf = clf.predict(x_test)

clf_report_cat = classification_report(y_test, pred_clf)
print(f"Classification Report : \n{clf_report_cat}")

clf_acc = accuracy_score(pred_clf, y_test)
print('DecisionTreeClassifier Accuracy Score: ', clf_acc)
```

**Model Comparison** 
 CatBoost and XgBoost comes close with 0.999 both, but **CatBoost** gives higher accuracy with 99.94%

```python
models = pd.DataFrame({
    'Model' : ['Logistic Regression','K Neighbors Classifier', 
               'XgBoost',  'CatBoost','Decision Tree Classifier'],

    'Score' : [lr.score(x_test,y_test),knn.score(x_test,y_test), xgb.score(x_test,y_test), cat.score(x_test,y_test) 
               , clf.score(x_test,y_test)]
})

models.sort_values(by = 'Score', ascending = False)
```

**Feature Importance**
    The variables which had an important impact on the model; *deposit type, more canceled than non canceled, required parking space, assigned room choice, market segment, total number of special requests* etc
    
Deposit type makes the guest think twice of cancelling their reservation. While guest who has more experience in cancelling their reservation before tend to do so again. Special request such as parking space provided and being assigned to room's choice also gives impact on cancellation trend.
